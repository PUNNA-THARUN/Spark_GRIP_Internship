# Each team Boundaries(Fours) and Total Runs scored by Fours 

```python
fours = df[df['batsman_runs']==4]
fours.groupby('batting_team')['batsman_runs'].agg([('runs by fours','sum'),('fours','count')])
```

```python
fours_batsman = fours.groupby('batsman')['batsman_runs'].agg([('runs_by_fours','sum'),('fours','count')])
top_batsman_fours = fours_batsman.sort_values("fours",ascending=0).iloc[:10,:]
top_batsman_fours
```

```python
# Plotting the Top 10 of boundaries scored
plt.figure(figsize=(15,8))
sns.barplot(x= top_batsman_fours.index, y= top_batsman_fours.fours, palette="Paired")
plt.xticks(rotation=90)
plt.xlabel('PLAYERS',fontsize= 15)
plt.ylabel('No. OF FOURS', fontsize= 15)
plt.title('TOP 10 PLAYERS SCORED HIGHEST FOURS', fontsize= 20)
plt.show()
```

# Each team Sixes and Total Runs scored by Sixes 

```python
six = df[df['batsman_runs'] == 6]
six.groupby('batting_team')['batsman_runs'].agg([('runs_by_six','sum'),('sixs','count')])
```

```python
six_batsman = six.groupby('batsman')['batsman_runs'].agg([('runs_by_six','sum'),('sixs','count')])
top_batsman_sixs = six_batsman.sort_values("sixs",ascending=0).iloc[:10,:]
top_batsman_sixs
```

```python
# Plotting the Top 10 Players of Sixs scored
plt.figure(figsize=(15,8))
sns.barplot(x= top_batsman_sixs.index, y= top_batsman_sixs.sixs, palette="Paired")
plt.xticks(rotation=90)
plt.xlabel('PLAYERS',fontsize= 15)
plt.ylabel('No. OF SIXES', fontsize= 15)
plt.title('TOP 10 PLAYERS SCORED HIGHEST SIXES', fontsize= 20)
plt.show()
```

```python
h_s = df.groupby('batsman')['batsman_runs'].agg(['sum'])
high_scorer = h_s.sort_values('sum', ascending= False).iloc[:10,:]
high_scorer
```

```python
# Plotting the Top 10 Highest Scorers of IPL
plt.figure(figsize=(15,8))
sns.barplot(x= high_scorer.index, y= high_scorer['sum'].values, palette="Paired")
plt.xticks(rotation=90)
plt.xlabel('PLAYERS',fontsize= 15)
plt.ylabel('HIGHEST RUNS', fontsize= 15)
plt.title('TOP 10 PLAYERS WITH HIGHEST RUNS IN IPL', fontsize= 20)
plt.show()
```

```python
bowl = df.groupby('bowler')['dismissal_kind'].agg(['count'])
high_wickets = bowl.sort_values('count', ascending= False).iloc[:10,:]
high_wickets
```

```python
# Plotting the Top 10 Wicket Takers of IPL
plt.figure(figsize=(15,8))
sns.barplot(x= high_wickets.index, y= high_wickets['count'].values, palette="Paired")
plt.xticks(rotation = 45)
plt.xlabel('PLAYERS',fontsize= 15)
plt.ylabel('HIGHEST WICKETS', fontsize= 15)
plt.title('TOP 10 PLAYERS WITH HIGHEST WICKETS IN IPL', fontsize= 20)
plt.show()
```
